```python
# for n= 29
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('interpolation_data.csv')

# Extract x and y values
x = data['x']
exact_y = data['exact_y']
linear = data['linear']
quadratic = data['quadratic']
cubic = data['cubic']

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the exact function
plt.plot(x, exact_y, 'k-', label='Exact y = sin(x^2)', linewidth=2)

# Plot the interpolations
plt.plot(x, linear, 'r--', label='Linear Interpolation', linewidth=1.5)
plt.plot(x, quadratic, 'b-.', label='Quadratic Interpolation', linewidth=1.5)
plt.plot(x, cubic, 'g:', label='Cubic Interpolation', linewidth=1.5)

# Add titles and labels
plt.title('Comparison of Interpolation Orders with Exact Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('/Users/minagerges/Downloads/interpolation_data (1).csv')

# Extract the columns
x = data['x']
original = data['Original']
linear = data['Linear']
PolyNomial = data['2nd_PolyNomial']

# Create the plot
plt.figure(figsize=(10, 6))

# Plot Original Function
plt.plot(x, original, label='Original Function: sin(x^2)', color='blue')

# Plot Linear Interpolation
plt.plot(x, linear, label='Linear Interpolation', linestyle='--', color='red')

# Plot Quadratic Interpolation
plt.plot(x, PolyNomial, label='2nd_Order Polynomial', linestyle=':', color='green')

# Add labels and title
plt.title('Interpolation of sin(x^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.xlim(0, 5.5)  # Set x limits
plt.ylim(-1.5, 2)  # Set y limits

# Show the plot
plt.show()


```
